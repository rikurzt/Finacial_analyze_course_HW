#證交所API

```python
import pandas as pd
import requests
import json
from datetime import datetime
```

```python
def call_twse_api(id,data=None, headers=None):
  today = datetime.today().strftime('%Y%m%d')
  stock_id = id
  url="https://www.twse.com.tw/rwd/zh/afterTrading/STOCK_DAY?date={today}&stockNo={stock_id}&response=json".format(today=today, stock_id=stock_id)

  try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    return response

  except requests.exceptions.RequestException as e:
    print(f"Error calling API: {e}")
    return None
```

```python
respond = call_twse_api("0050")
respond.json()['data']
```

```python
df = pd.DataFrame(respond.json().get('data'))
df.columns = respond.json().get('fields')
df
```

# Yahoo爬蟲

```python
from bs4 import BeautifulSoup
```

```python
stock_id = "2330"
url = "https://tw.stock.yahoo.com/quote/{stock_id}.TW".format(stock_id=stock_id)
response = requests.get(url)
html_content = response.text
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
table = soup.find('div', {'class': 'Fx(n) W(316px) Bxz(bb) Pstart(16px) Pt(12px)'})
rows = table.find_all('li')


data = []

for row in rows:
    cells_col = row.find_all('span')
    data.append([cell.text.strip() for cell in cells_col])

data = list(zip(*data))
data
```

```python
df = pd.DataFrame(data[1:], columns=data[0])
df
```

## 取得月報表

```python
stock_id = "2330"
url = "https://tw.stock.yahoo.com/quote/{stock_id}.TW/income-statement".format(stock_id=stock_id)
response = requests.get(url)
html_content = response.text
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
table = soup.find('div', {'class': 'Pos(r) Ov(h)'})
header = table.find('div', {'class': 'table-header Ovx(s) Ovy(h) W(100%)'})
rows = table.find('div', {'class': 'table-body-wrapper'})

data = []



for row in rows.find_all('li'):
    cells_col = row.find_all('span')
    data.append([cell.text.strip() for cell in cells_col])


```

```python
df = pd.DataFrame(data, columns=header.text.split())
df
```
